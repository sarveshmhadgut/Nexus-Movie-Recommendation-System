```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data.title_year.value_counts(dropna=False).sort_index().plot(
    kind="barh", figsize=(15, 16), color="cyan", edgecolor="black"
)
plt.show()
```

```python
data = data.loc[
    :,
    [
        "director_name",
        "actor_1_name",
        "actor_2_name",
        "actor_3_name",
        "genres",
        "movie_title",
        "imdb_score",
        "plot_keywords",
    ],
]
```

```python
data.shape
```

```python
data.head(5043)
```

```python
data["actor_1_name"] = data["actor_1_name"].replace(np.nan, "unknown")
data["actor_2_name"] = data["actor_2_name"].replace(np.nan, "unknown")
data["actor_3_name"] = data["actor_3_name"].replace(np.nan, "unknown")
data["director_name"] = data["director_name"].replace(np.nan, "unknown")
```

```python
data
```

```python
data["genres"] = data["genres"].str.replace("|", " ")
```

```python
data["plot_keywords"] = data["plot_keywords"].str.replace("|", " ")
data["plot_keywords"] = data["plot_keywords"].fillna("unknown")
```

```python
data
```

```python
data["movie_title"] = data["movie_title"].str.lower()
data["movie_title"] = data["movie_title"].apply(lambda x: x[:-1])
```

```python
data.to_csv("../datasets/data.csv", index=False)
```
